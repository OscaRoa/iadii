### Modelo de Retención de Memoria

En este cuadernillo reproducimos dos versiones del modelo de retención de memoria discutido en Shiffrin et al. (2008) y en el capítulo 10 de Lee y Wagenmakers (2013).

En un experimento hipotético, tres participantes estudiaron una lista de 18 palabras, e intentaron recordarlas después de diferentes intervalos de tiempo. La cantidad de palabras recordadas por cada participante después de cada intervalo se presenta en el siguiente formato:

| Subject | 1 | 2 | 4 | 7 | 12 | 21 | 35 | 59 | 99 | 200 |
|:-:|:--:|:--:|:--:|:--:|:-:|:-:|:-:|:-:|:-:|:-:|
| 1 | 18 | 18 | 16 | 13 | 9 | 6 | 4 | 4 | 4 | ? |
| 2 | 17 | 13 | 9  | 6  | 4 | 4 | 4 | 4 | 4 | ? |
| 3 | 14 | 10 | 6  | 4  | 4 | 4 | 4 | 4 | 4 | ? |
| 4 | ?  | ?  | ?  | ?  | ? | ? | ? | ? | ? | ? |

El ejercicio de modelamiento consiste en intentar explicar estas observaciones utilizando un modelo de memoria de *decaimiento exponencial*.

En pocas palabras, el modelo supone que la *probabilidad* de recordar cada palabra es una función exponencial del tiempo transcurrido desde la memorización hasta el momento de recordarla:

$$\theta\gets exp(-\alpha t)+\beta$$

$$\theta\gets e^{-\alpha t}+\beta$$

En donde la variable $t$ representa el tiempo de recuerdo y los parámetros no-observables $\alpha$ ("tasa de decaimiento") y $\beta$ ("línea base de recuerdo") controlan diferentes aspectos de la función. Se recomienda consultar las fuentes originales para mayores detalles del modelo, así como estudiar el comportamiento de esta función central bajo diferentes valores paramétricos, como se muestra a continuación:

```r
rm(list=ls())
# Analyzing the core model
tm <- c(1,2,4,7,12,21,35,59,99,200)

alpha_1 <- 0
beta_1 <- 0
th_1 <- exp(-alpha_1*tm)+beta_1
col_1 <- '#ee0000'

alpha_2 <- 1
beta_2 <- 0.4
col_2 <- '#0000ee'
th_2 <- exp(-alpha_2*tm)+beta_2

alpha_3 <- .5
beta_3 <- .4
col_3 <- '#00cc00'
th_3 <- exp(-alpha_3*tm)+beta_3

alpha_4 <- 0.01
beta_4 <- 0.05
col_4 <- '#ee8800'
th_4 <- exp(-alpha_4*tm)+beta_4


options(repr.plot.width = 8, repr.plot.height = 8) # Plot size in notebook
par(mar=rep(5,4))
plot(NULL,ylim=c(0,1),xlim=c(1,length(tm)),axes=F,ann=F)
lines(1:length(tm),th_1,col=col_1,lwd=4,type='o',pch=16,cex=1.5)
lines(1:length(tm),th_2,col=col_2,lwd=4,type='o',pch=16,cex=1.5)
lines(1:length(tm),th_3,col=col_3,lwd=4,type='o',pch=16,cex=1.5)
lines(1:length(tm),th_4,col=col_4,lwd=4,type='o',pch=16,cex=1.5)
axis(1,at=1:length(tm),labels=tm,cex.axis=1.5,padj=0.25)
axis(2,cex.axis=1.5,las=1)
mtext('Time Lags',1,cex=1.5,line=3)
mtext('Theta (probability of recalling)',2,cex=1.5,line=3.5)
legend(2,.3,
       legend=c(paste('alpha=',alpha_1,' beta=',beta_1),
               paste('alpha=',alpha_2,' beta=',beta_2),
               paste('alpha=',alpha_3,' beta=',beta_3),
               paste('alpha=',alpha_4,' beta=',beta_4)),
       cex=1.25,
       col=c(col_1,col_2,col_3,col_4),
       text.width=3.6,
       bg='#ffffffcc',
       lwd=3,pch=16)
```

En tanto que las características de memoria $\alpha$ y $\beta$ no son observables, pero su valor es relevante respecto de cierta teoría psicológica, es posible calcular sus valores más probables a la luz de las observaciones del experimento.

Las células siguientes muestran cómo calcular dicha incertidumbre y reproducen los resultados publicados previamente.

```r
rm(list=ls())

# Observed
n_targets <- 18 # Number of words presented to participants
time <- c(1,2,4,7,12,21,35,59,99,200) # Time lags after which participants recall
n_times <- length(time) # Number of time lags
n_subjs <- 4 # Number of subject

n_recalled <- array(dim=c(n_subjs,n_times)) # Number of words recalled by each participant after each lag
n_recalled[1,] <- c(18,18,16,13,9,6,4,4,4,NA)
n_recalled[2,] <- c(17,13,9,6,4,4,4,4,4,NA)
n_recalled[3,] <- c(14,10,6,4,4,4,4,4,4,NA)
n_recalled[4,] <- NA

data_jags <- list('n_targets','time','n_times','n_subjs','n_recalled') # Everything is going to JAGS

n_recalled
```

```r
# Unobserved (in this example these labels are consistent across models)
unobs <- c('alpha_post','beta_post','n_recalled_post',
           'alpha_prior','beta_prior','n_recalled_prior',
          'mu_alpha_prior','mu_alpha_post',
          'mu_beta_prior','mu_beta_post',
          'lambda_beta_prior','lambda_beta_post',
          'lambda_alpha_prior','lambda_alpha_post')
```

```r
# Model 1: no individual differences
write('

model{
  
  alpha_prior~dbeta(1,1)
  beta_prior~dbeta(1,1)
  alpha_post~dbeta(1,1)
  beta_post~dbeta(1,1)
    
 for(j in 1:n_times){
    theta_prior[j] <- min(.99,exp(-alpha_prior*time[j])+beta_prior)
    theta_post[j] <- min(.99,exp(-alpha_post*time[j])+beta_post)
    #theta_post[j] <- exp(-alpha_post*time[j])+beta_post


    for(i in 1:n_subjs){      
      n_recalled_prior[i,j]~dbin(theta_prior[j],n_targets)
      n_recalled[i,j]~dbin(theta_post[j],n_targets)
      n_recalled_post[i,j]~dbin(theta_post[j],n_targets)
    }
  }
  
}
','no_individual_differences.bug')

# Inference over Model 1
library('R2jags')
bayes_no_ind_diff <- jags(data=data_jags,
                          parameters.to.save=unobs,
                          model.file='no_individual_differences.bug')
nds_no_ind_diff <- bayes_no_ind_diff$BUGSoutput$sims.list
```

```r
# Model 2: non-structured individual differences
write('

model{
  
  for(i in 1:n_subjs){
    alpha_prior[i]~dbeta(1,1)
    beta_prior[i]~dbeta(1,1)
    alpha_post[i]~dbeta(1,1)
    beta_post[i]~dbeta(1,1)
    
    for(j in 1:n_times){
      theta_prior[i,j] <- min(.99,exp(-alpha_prior[i]*time[j])+beta_prior[i])
      theta_post[i,j] <- min(.99,exp(-alpha_post[i]*time[j])+beta_post[i])

      n_recalled_prior[i,j]~dbin(theta_prior[i,j],n_targets)
      n_recalled[i,j]~dbin(theta_post[i,j],n_targets)
      n_recalled_post[i,j]~dbin(theta_post[i,j],n_targets)
    }
  }
  
}
','individual_differences.bug')

# Inference over Model 2
bayes_ind_diff <- jags(data=data_jags,
                       parameters.to.save=unobs,
                       model.file='individual_differences.bug')
nds_ind_diff <- bayes_ind_diff$BUGSoutput$sims.list
```

```r
# Model 3: Hierarchical individual differences
write('

model{
  
  # Parents of Decay Rate (alpha) 
  mu_alpha_post~dbeta(1,1)
  mu_alpha_prior~dbeta(1,1)
  lambda_alpha_post~dgamma(0.001,0.001)T(0.01,)
  lambda_alpha_prior~dgamma(0.001,0.001)T(0.01,)
    
  # Parents of Baseline (beta)
  mu_beta_post~dbeta(1,1)
  mu_beta_prior~dbeta(1,1)
  lambda_beta_post~dgamma(0.001,0.001)T(0.01,)
  lambda_beta_prior~dgamma(0.001,0.001)T(0.01,)
  
  for(i in 1:n_subjs){
    alpha_prior[i]~dnorm(mu_alpha_prior,lambda_alpha_prior)T(0,1)
    alpha_post[i]~dnorm(mu_alpha_post,lambda_alpha_post)T(0,1)
    beta_prior[i]~dnorm(mu_beta_prior,lambda_beta_prior)T(0,1)
    beta_post[i]~dnorm(mu_beta_post,lambda_beta_post)T(0,1)
    
    for(j in 1:n_times){
      theta_prior[i,j] <- min(.99,exp(-alpha_prior[i]*time[j])+beta_prior[i])
      theta_post[i,j] <- min(.99,exp(-alpha_post[i]*time[j])+beta_post[i])

      n_recalled_prior[i,j]~dbin(theta_prior[i,j],n_targets)
      n_recalled[i,j]~dbin(theta_post[i,j],n_targets)
      n_recalled_post[i,j]~dbin(theta_post[i,j],n_targets)
    }
  }
  
}
','hier_individual_differences.bug')

# Inference over Model 3
bayes_hier_ind_diff <- jags(data=data_jags,
                       parameters.to.save=unobs,
                       model.file='hier_individual_differences.bug',
                       n.iter=8000,
                       n.burnin=1000)
nds_hier_ind_diff <- bayes_hier_ind_diff$BUGSoutput$sims.list
```

```r
# Model 4: semi-structured individual differences
write('

model{
  
  beta_prior~dbeta(1,1)
  beta_post~dbeta(1,1)
  
  for(i in 1:n_subjs){
    alpha_prior[i]~dbeta(1,1)
    alpha_post[i]~dbeta(1,1)
    
    for(j in 1:n_times){
      theta_prior[i,j] <- min(.99,exp(-alpha_prior[i]*time[j])+beta_prior)
      theta_post[i,j] <- min(.99,exp(-alpha_post[i]*time[j])+beta_post)

      n_recalled_prior[i,j]~dbin(theta_prior[i,j],n_targets)
      n_recalled[i,j]~dbin(theta_post[i,j],n_targets)
      n_recalled_post[i,j]~dbin(theta_post[i,j],n_targets)
    }
  }
  
}
','semi_individual_differences.bug')

# Inference over Model 4
bayes_s_ind_diff <- jags(data=data_jags,
                       parameters.to.save=unobs,
                       model.file='semi_individual_differences.bug')
nds_s_ind_diff <- bayes_s_ind_diff$BUGSoutput$sims.list
```
